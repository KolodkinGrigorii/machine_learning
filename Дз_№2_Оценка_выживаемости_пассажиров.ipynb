Требуется предсказать бинарный признак `Survived` по остальным признакам. Это *задача восстановления регрессии*.

Столбец `Cabin` содержит много пропущенных значений, что делает проблематичным его заполнение. Поэтому целесообразней его удалить, т.к. маловероятно, что он даст ценную информацию.
```python
data_raw = data_raw.drop('Cabin', axis=1)

mode_age = data_raw['Age'].mode()[0]
data_raw['Age'].fillna(mode_age, inplace=True)

mode_embarked = data_raw['Embarked'].mode()[0]
data_raw['Embarked'].fillna(mode_embarked, inplace=True)

mode_fare = data_raw['Fare'].mode()[0]
data_raw['Fare'].fillna(mode_fare, inplace=True)
```

## 5. Разбивка на тестовую и обучающую выборки

Для предсказания выживаемости будем использовать все признаки, кроме имени, PassengerId, Ticket, Embarked, а также преобразуем значения пола male/female в бинарные 0 и 1:


```python
data_raw['Sex'] = data_raw['Sex'].map({'male': 1, 'female': 0})
X = data_raw.drop(['Survived', 'Name', 'PassengerId', 'Ticket', 'Embarked'], axis=1)
y = data_raw['Survived']
```

Разобьем данные на обучающую и тестовую выборки в пропорции 2:1 (68% - обучающая выборка, 32% - тестовая):



```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.32, random_state = 42)

N_train, _ = X_train.shape
N_test,  _ = X_test.shape

N_train, N_test
```

## 6. Запуск классификатора

Для данного датасета нормализация не нужна, потому что признаки имеют сопоставимые диапазоны: Age (0-80), Fare (0-512), SibSp/Parch (0-8). Кроме этого, большинство признаков уже категориальные.

Исходя из вышесказанного, в качестве классификатора стоит взять Random Forest/Gradient Boosting, потому что:
1. Отлично работают со смешанными данными - в Titanic есть и числа (возраст, цена), и категории (пол, порт)

2. Не требуют нормализации - сами справляются с разными масштабами признаков

Для примера возьмем Random Forest, но перед этим преобразуем признак `Sex` в бинарное значение 0 и 1

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=200) # по умолчанию n_estimators=100
model.fit(X_train, y_train)
```

```python
y_train_predict = model.predict(X_train)
y_test_predict  = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

RMSE_train = mean_squared_error(y_train, y_train_predict)**.5
RMSE_test  = mean_squared_error(y_test,  y_test_predict)**.5
R2_train = r2_score(y_train, y_train_predict)
R2_test  = r2_score(y_test,  y_test_predict)
```

```python
RMSE_table = pd.DataFrame(columns=['Train RMSE', 'Test RMSE'])
RMSE_table.loc['Random Forest', :] = (RMSE_train, RMSE_test)
RMSE_table
```

Как мы видим, обучающая модель демонстрирует достаточно низкую ошибку, что говорит о хорошем соответствии модели обучающим данным. Говоря о тестовой выборке, то у неё ошибка немного выше, что свидетельствует о некоторой потере точности при обобщении на новые данные. Это нормально, поскольку модель может немного "подгоняться" под обучающие данные.

```python
R2_table = pd.DataFrame(columns=('train', 'test'))
R2_table.loc['Random Forest', :] = (R2_train, R2_test)
R2_table
```

Теперь найдем оптимальный гиперпараметр и обучим модель при этом оптимальном значении

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import PolynomialFeatures
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_selector
!pip install feature_engine
from feature_engine.selection import DropFeatures
```

```python
ppl = Pipeline([
    # Заполнение пропущенных значений:
    ('impute', ColumnTransformer([
        # Заполнение пропущенных числовых признаков:
        ('num', make_pipeline(
            SimpleImputer(strategy='median'),
            StandardScaler()),
            make_column_selector(dtype_include='float')
        ),
        # Заполнение пропущенных категориальных признаков:
        ('cat', make_pipeline(
            SimpleImputer(strategy='most_frequent'),
            OneHotEncoder()),
            make_column_selector(dtype_include='category')
        )])
    ),
    # Обучение:
    ('fit', RandomForestRegressor())
])
```

```python
ppl.fit(X_train, y_train)
y_train_predict = ppl.predict(X_train)
y_test_predict = ppl.predict(X_test)
RMSE_train = mean_squared_error(y_train, y_train_predict)**.5
RMSE_test  = mean_squared_error(y_test,  y_test_predict)**.5
RMSE_train, RMSE_test
```

Получили значение ошибок у обучающей выборки 0.39 и 0.52 у тестовой
